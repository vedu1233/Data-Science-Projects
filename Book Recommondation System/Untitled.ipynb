# Build a popularity Basesd Recommendation system

```python
# reveal the book who has highest rating 
```

```python
ratings_with_name1 = ratings.merge(books,on ="ISBN")
```

```python
ratings_with_name1
```

```python
num_rating_df1 =ratings_with_name1.groupby('Book-Title').count()["Book-Rating"].reset_index()
```

```python
num_rating_df1.rename(columns ={"Book-Rating":"num_ratings"},inplace = True)
```

```python
num_rating_df1
```

```python
avg_rating_df = ratings_with_name1.groupby('Book-Title')['Book-Rating'].mean().reset_index()
```

```python
popularity_df  = num_rating_df1.merge(avg_rating_df ,on ="Book-Title")
popularity_df
```

```python
popular_df =popularity_df[popularity_df["num_ratings"] >=250].sort_values('avg_rating',ascending =False).head(50)
```

```python
popular_df
```

```python
popular_df =popular_df.merge(books,on ="Book-Title").drop_duplicates('Book-Title')
```

```python
popular_df =popular_df[["Book-Title","Book-Author_x","Image-URL-M_x","num_ratings","avg_rating"]]
```

```python
popular_df # this is our popularity based reccommnedation system
```

```python
import pickle 
```

```python
file = open("popular_df.pkl","wb")

pickle.dump(popular_df,file)
```

# Collborative recommendation System

```python
ratings_with_name1
# user id that rate number of books
```

```python
ratings_with_name1.groupby("User-ID").count()
# user id that rate number of books
```

```python
ratings_with_name1.groupby("User-ID").count()["Book-Rating"]>200
# user id that rate number of books
```

```python
x = ratings_with_name1.groupby("User-ID").count()["Book-Rating"]>200# user that reads more than 200 boooks and gives ratings
# user id that rate number of books
# user id 2 rate 1 books
# user id 8 rate 17 books
x[x].index
```

```python
padhe_likhe_users =x[x].index
```

```python
padhe_likhe_users
```

```python
filtered_rating =ratings_with_name1[ratings_with_name1["User-ID"].isin(padhe_likhe_users)]
```

```python
filtered_rating # returns the user id's whos make revies more than 200 books
```

```python
y =filtered_rating.groupby('Book-Title').count()["Book-Rating"]>=50
famous_books =y[y].index
```

```python
famous_books
```

```python
y
```

```python
famous_books
```

```python
final_ratings =filtered_rating[filtered_rating["Book-Title"].isin(famous_books)]
```

```python
final_ratings
```

```python
pt = final_ratings.pivot_table(index ="Book-Title",columns ='User-ID',values ='Book-Rating' )
# user who read more than 200 books and 50 popular books
```

```python
pt.fillna(0,inplace = True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score =cosine_similarity(pt)
```

```python
similarity_score[0]
#1 means similarity scores of 1984 from others points
```

```python
np.where(pt.index =='1984')[0][0] # [0] if for array index second[0] is for real index
```

```python
np.where(pt.index =='4 Blondes')[0][0] # index of book name =4 Blondes
```

```python
def recommend(book_name): # here we define the function
    # fetch index name using book name
    index = np.where(pt.index==book_name)[0][0] # returns the index of book 
    similar_items =sorted(list(enumerate(similarity_score[index])),key =lambda x:x[1],reverse =True)[1:6] 
    
    for i in similar_items:
        print(pt.index[i[0]])
   # return suggestion
```

```python
sorted(list(enumerate(similarity_score[0])),key =lambda x:x[1],reverse =True)[1:6]
```

```python
recommend('1984')
#the movies similiar to movie ="1984" will recommonded 
# only [1:6] similiarity score based on descending order will dilsay
```

```python
recommend('Message in a Bottle')
```
